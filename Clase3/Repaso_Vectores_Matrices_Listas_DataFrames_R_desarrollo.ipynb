## 4) Listas
Empaquetemos resúmenes y objetos útiles en una **lista** para guardar un \"informe\" sencillo. Por ejemplo,
- Resumen de las variables en matriz antes creada
- El o los automóviles con mayor mpg
- El o los automóviles con mayor hp


```r
resultado_orden_descendente_mpg <- sort(Matriz_var[, 'mpg'], index.return = T, decreasing = TRUE)

resultado_orden_descendente_hp <- sort(Matriz_var[, 'hp'], index.return = T, decreasing = TRUE)



list(resumen = summary(Matriz_var),
      mayor_mpg = rownames(head( Matriz_var[resultado_orden_descendente_mpg[[2]], ], 3)),
      mayor_hp = rownames(head( Matriz_var[resultado_orden_descendente_hp[[2]], ], 3)))
```
